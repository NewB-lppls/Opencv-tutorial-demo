```python
import cv2
```

```python
# Hiển thị hình ảnh
img = cv2.imread('Capture.PNG')

# Lấy kích thước hình ảnh
(h,w,d) = img.shape
print(f'Chiều cao: {h}, Chiều rộng: {w}, Chiều dài ảnh: {d}\n')

# Show hình ảnh
from google.colab.patches import cv2_imshow
cv2_imshow(img)
img = cv2.resize(img,(300,300))
```

```python
# Đọc ảnh gray
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
x = cv2.resize(img_gray, (200,200))
cv2_imshow(x)
```

```python
#Lưu ảnh
cv2.imwrite('ảnh.PNG',x)
```

```python
# Phương pháp cắt ảnh tương tự như slice mảng trong numpy
x = img[80:110,80:200]
cv2_imshow(x)
```

```python
# Thay đổi kích thước ảnh
x = cv2.resize(img, (200,200))
cv2_imshow(x)
```

```python
# xoay ảnh với điểm xoay ở tâm điểm tọa độ x100,y100 và xoay 45 độ
# xác định gốc xoay và hướng xoay
ra = cv2.getRotationMatrix2D((100,100), 45, 1.0)

rotated = cv2.warpAffine(img, ra, (200, 200))
cv2_imshow(rotated)
```

```python
#vẽ đường thẳng trong opencv
# Tạo ảnh bằng numpy
import numpy as np
x = np.random.randint(255, size =(200,200),dtype=np.uint8)

# tọa độ bắt đầu, kết thúc, chiều rộng đường line

id_star = (50,50)
id_end =(100,100)
thinkness = 10
# vẽ đường line (0,0,0) là vùng màu
x1 = cv2.line(x,id_star,id_end,(0,0,0),thinkness)
cv2_imshow(x1)
```

```python
# vẽ đường tròn cv2.circle 
center = (100,100)
radius = 30
color =(0,0,255)
thinkness = 7
x1 = cv2.circle(img, center, radius, color, thinkness)
cv2_imshow(x1)
```

```python
# vẽ hình chữ nhật
x2 = cv2.rectangle(img,id_star,id_end,color,thinkness)
cv2_imshow(x2)
```

```python
# viết chữ lên ảnh
text = 'hello world'
org = (100,100)
font = cv2.FONT_HERSHEY_SIMPLEX
fontscale = 1
color= (0,0,0)
thinkness = 3
x1 = cv2.putText(img, text, org, font, fontscale,color,thinkness)

cv2_imshow(x1)
```
